```python
#Adding location of the "club"
def getCanton(clubName,dictionnaryLocality=None,dictionnaryRegion=None,SwissSafety=True):
    '''Returns the location of the club with a tuple (lat,lng)'''
    
    clubName = clubName.replace(";",",")
    region=None
    locality = None
    lat = None
    lng = None
    
    if(dictionnaryRegion != None and (clubName in dictionnaryRegion)):
        region = dictionnaryRegion[clubName]
    
    if(dictionnaryLocality != None and (clubName in dictionnaryLocality)):
        locality = dictionnaryLocality[clubName]
    
    
    if(locality !=None and region!=None):
        return region,locality
    else:
        if(locality==None): #Try with geoCodeAPI
            baseURL1 = 'https://maps.googleapis.com/maps/api/place/textsearch/json'
            try:
                region,locality = getFromAPI(clubName,baseURL1)
            except:
                region,locality  = None,None
            
        if(locality==None): #Try with textSearchAPI
            baseURL2 = 'https://maps.googleapis.com/maps/api/geocode/json'
            try:
                region,locality  = getFromAPI(clubName,baseURL2)
            except:
                region,locality  = None,None
                
        #If nothing found     
        if(region==None and locality==None):
            address_comp = list(clubName.split(","))[::-1]
            
            #timeout
            timeout = 5
            i = 0
            if(len(address_comp)>1):
                for ac in address_comp:
                    i+=1
                    region,locality = getCanton(ac,None,None)
                    if(region!=None):
                        break
                    if(i>timeout):
                        break
                
                
        if(dictionnaryRegion!=None):
            dictionnaryRegion.update({clubName : region})
        if(dictionnaryLocality!=None):
            dictionnaryLocality.update({clubName : locality})
            
    return region, locality
```

```python
def completeRegionAndLocality(dataframe):
    dataframe['region'] = None
    dataframe = dataframe[["place" ,"club" ,"address","location","region","event" ,
                           "date","artists","genre","coordinates"]]
    
    AddressDic = {}
    LocalityDic = {}
    
    i = 0
    print(dataframe.shape)
    
    for id,row in dataframe.iterrows():
        
        i+=1
        if(i%200==0):
            print(i)
        
        place = row.place
        address = row.address
        location = row.location
       
        region = None
        locality = None
        #Replace invalid location
        if(":" in str(location) or "{" in str(location) or "nan" in str(location).lower()):
            location = None
            
        noneLocation = (location == None or str(location)=="nan")    
            
        if(noneLocation and address!=None):
            region,locality = getCanton(address,AddressDic,LocalityDic)
            
        elif(noneLocation == None and place!=None):
            region,locality = getCanton(place,AddressDic,LocalityDic)
            
        elif(location != None):
            region,locality = getCanton(location,AddressDic,LocalityDic)
            
        dataframe.loc[id,"location"] = locality
        dataframe.loc[id,"region"] = region
        
    return dataframe
```

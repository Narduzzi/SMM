## Interacting with the data

```python
import matplotlib.pyplot as plt


Artist = "Getter"
spotify_g = getGenresFromSpotify(Artist)
ra_g = getGenresFromRA(Artist)
genres_tot = spotify_g+ra_g

my_norm = matplotlib.colors.Normalize(0, 6)
my_cmap = cm.get_cmap('gist_rainbow')

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
dictGenres = createDictionnary()

genres = [dictGenres.get(x) for x in genres_tot]#DF["Genres"][0]]
labels = set(genres)
sizes = [genres.count(x) for x in labels]
explode = [0 for x in labels]  # only "explode" the 2nd slice (i.e. 'Hogs')

colors = [1+i for i in range(0,len(labels))]

fig1, ax1 = plt.subplots(figsize=(6,6))
ax1.pie(sizes, explode=explode, labels=labels,colors=my_cmap(my_norm(colors)), autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()

genres = [x for x in genres_tot]#DF["Genres"][0]]
labels = set(genres)
sizes = [genres.count(x) for x in labels]
explode = [0 for x in labels]  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots(figsize=(6,6))
ax1.pie(sizes, explode=explode, labels=labels,colors=my_cmap(sizes), autopct='%1.1f%%',
        shadow=True, startangle=90,)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
